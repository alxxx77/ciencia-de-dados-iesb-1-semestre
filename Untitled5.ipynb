```python
n1 = float(input("Nota 1: "))
n2 = float(input("Nota 2: "))
n3 = float(input("Nota 3: "))
n4 = float(input("Nota 4: "))

media = (n1 + n2 + n3 + n4) / 4
print(f"A média é {media}")
```

```python
metros = float(input("Digite o valor em metros: "))
centimetros = metros * 100
print(f"{metros} metros = {centimetros} centímetros")
```

```python
import math

raio = float(input("Digite o raio do círculo: "))
area = math.pi * raio**2
print(f"A área do círculo é {area}")
```

```python
lado = float(input("Digite o lado do quadrado: "))
area = lado ** 2
dobro = area * 2

print(f"A área é {area}")
print(f"O dobro da área é {dobro}")
```

```python
valor_hora = float(input("Quanto você ganha por hora? "))
horas = float(input("Quantas horas trabalhou no mês? "))

salario = valor_hora * horas
print(f"Seu salário este mês foi de R$ {salario:.2f}")
```

```python
f = float(input("Digite a temperatura em Fahrenheit: "))
c = 5 * ((f - 32) / 9)

print(f"{f}°F equivale a {c:.2f}°C")
```

```python
c = float(input("Digite a temperatura em Celsius: "))
f = (c * 9/5) + 32

print(f"{c}°C equivale a {f:.2f}°F")
```

```python
a = int(input("Digite o primeiro número inteiro: "))
b = int(input("Digite o segundo número inteiro: "))
c = float(input("Digite um número real: "))

produto = (2 * a) * (b / 2)
soma = (3 * a) + c
cubo = c ** 3

print(f"Produto: {produto}")
print(f"Soma: {soma}")
print(f"Cubo: {cubo}")
```

```python
altura = float(input("Digite sua altura (em metros): "))
peso_ideal = (72.7 * altura) - 58

print(f"Seu peso ideal é {peso_ideal:.2f} kg")
```
