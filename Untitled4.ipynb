```python
def calcula_desconto(valor, desconto):
    valor_final = valor - (valor * desconto / 100)
    return valor_final
resultado = calcula_desconto(100, 10)
print(resultado)
```

```python
def classifica_nota(nota):
    if nota >= 7:
        return "Aprovado"
    elif nota >= 5:
        return "Recuperação"
    else:
        return "Reprovado"
resultado = classifica_nota(6.5)
print(resultado)
```

```python
numeros = [10, -5, 0, 7, -2, 0, 3]

positivos = 0
negativos = 0
zeros = 0

for numero in numeros:
    if numero > 0:
        positivos += 1
    elif numero < 0:
        negativos += 1
    else:
        zeros += 1
print("Positivos:", positivos)
print("Negativos:", negativos)
print("Zeros:", zeros)
```
