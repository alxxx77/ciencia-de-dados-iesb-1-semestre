```python
def calcula_desconto(valor, desconto):
    return valor * (1 - desconto / 100)

resultado = calcula_desconto(100, 10)
print(resultado)
```

```python
def classifica_nota(nota):
    if nota >= 7:
        return "Aprovado"
    elif nota >= 5:
        return "Recuperação"
    else:
        return "Reprovado"

print(classifica_nota(6.5))
```

```python
numeros = [10, -3, 0, 5, -8, 0, 2]

positivos = 0
negativos = 0
zeros = 0

for n in numeros:
    if n > 0:
        positivos += 1
    elif n < 0:
        negativos += 1
    else:
        zeros += 1

print(f"Positivos: {positivos}")
print(f"Negativos: {negativos}")
print(f"Zeros: {zeros}")
```
